```python
import pandas_datareader.data as web
import datetime
from bokeh.plotting import figure, show, output_file
```

```python
start=datetime.datetime(2018,1,1)
end=datetime.datetime(2018,8,30)

symbol = 'WIKI/AAPL'
df = web.DataReader(symbol, data_source="quandl", start=start, end=end)
```

```python
twlo = web.DataReader('TWLO','iex',start,end)
```

```python
twlo
```

```python
p = figure(x_axis_type='datetime', width=1000, height=300, title="Lawrence Apple Candlestick Chart")
```

```python
df.index[df.Close > df.Open]
```

```python
hours_12 = 12*60*60*1000
```

```python
p.rect(df.index[df.Close > df.Open],(df.Open + df.Close)/2,hours_12,abs(df.Open-df.Close),fill_color="green", line_color="black")
p.rect(df.index[df.Close < df.Open],(df.Open + df.Close)/2,hours_12,abs(df.Open-df.Close),fill_color="red", line_color="black")

output_file("CS.html")
show(p)
```
